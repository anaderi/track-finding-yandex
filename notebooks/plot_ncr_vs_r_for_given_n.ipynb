# Exploring nCr

In studying combinatorics and probability, the combination function is everywhere.

The function : 
$$ nCr = \frac{n!}{r!\left(n-r\right)!} $$
is maximised for r = n/2

This plots this behaviour, exposing the shape of this maximum.  

For each $0 < n_i < N$, all viable r_j's, i.e. $0 < r_j <= n$, are cacluated.

Note that the image binning is a little lazy, with the lower left corner of the bin corresponding to the bin value

```python
import math
import numpy as np
%pylab inline

figsize(10,10)

# Define the combinatorial function
def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)

# Define our image size
nN = 100
# Initialize the array
an_array = np.zeros((nN, nN))

# Caculate all values
for n in range(0,nN):
    for r in range(0,n+1):
        an_array[n,r] = nCr(n,r)
# Normalize each row by the value of r that maximises that row
an_array = (an_array.astype(float).T / (np.amax(an_array, axis=1)).astype(float)).T
# Plot the image
plt.imshow(an_array, interpolation='nearest', origin='low',
           extent=[0, nN, 0, nN])
plt.colorbar()
plt.title("nCr / max_r(nCr) for a given n vs. (n,r)")
plt.xlabel("r")
plt.ylabel("n")
plt.show()
# Plot a slice at n=85
plt.title("nCr / max_r(nCr) for n = 50 vs. r")
plt.ylabel("nCr/max(nCr) for n=50")
plt.xlabel("r")
plt.plot(range(nN), an_array[50])
plt.show()
```
