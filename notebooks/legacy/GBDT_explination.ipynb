## Brief Example of How this Loss Function Works

### Code preamble

```python
# import our libraries
import numpy as np
%pylab inline
```

```python
# define the loss function as in sklearn.ensemble
def log_likelihood(y_1, y_2):
    return -2.0 * ((y_1 * y_2) - np.logaddexp(0.0, y_2))
```

```python
# Make an array of 50 points, evenly spaced in [0.0, 1.0]
# i.e. a "linearly spaced" array
dimension = 50
all_y1 = np.linspace(0, 1., num=dimension)
```

The key point here is that y_1, i.e. our "y", can only have two values : 

* y = 1  means signal
* y = 0  means background

BUT y_grad can have any value between [0.0, 1.0].  Ideally, y_grad = y, or is very close

### Ex 1.  y_1 = y = 1, i.e. signal hit
