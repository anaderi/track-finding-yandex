# Drawing

## Drawing Function(s)

```python
def setup_plot_for_detector(size=(10, 10)):
    figsize(*size)
    axes(polar=True)
    # set appropriate
    ylim(0, 85)   
```

## Drawing Sandbox

### Drawing the Signal Tracks

```python
setup_plot_for_detector()
scatter(phis_SG, rhos_SG, marker='.', s=5)

event = 18
print signal.n_events
SG_trk = where(signal.get_hit_types(event) != 0)
scatter(phis_SG[SG_trk], rhos_SG[SG_trk], marker='o')
```

```python
setup_plot_for_detector()
scatter(phis_SG, rhos_SG, marker='.', s=5)

event = 15
print mu_capt_bg.n_events
MuCBG_trk = where(mu_capt_bg.get_hit_types(event) != 0)
scatter(phis_MuCBG[MuCBG_trk], rhos_MuCBG[MuCBG_trk], marker='o')
```
